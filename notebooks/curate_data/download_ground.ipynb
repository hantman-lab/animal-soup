```python
import pandas as pd
from pathlib import Path
import os
from scipy.io import loadmat
from tqdm import tqdm
import shutil
```

```python
df = pd.read_hdf("/data/caitlin/full_hand_labels.hdf")
```

```python
df.iloc[1]['full_hand_labels']
```

```python
mat_files = Path('/data/caitlin/mat_files/')
```

```python
vid_paths = list()
for row in tqdm(df.iterrows()):
    m = loadmat(mat_files.joinpath(row[1]['mat_path']).with_suffix('.mat'))
    keys = row[1]['full_hand_labels'].keys()
    for k in keys:
        index = int(k.split('_v')[-1]) - 1
        try:
            vid_paths.append(m['data']['id'][index][0][0])
        except KeyError:
            try: 
                vid_paths.append(m['data1']['id'][index][0][0])
            except KeyError:
                continue
```
