```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
test_df = pd.read_hdf('/data/caitlin/cross_validation/k2/k2_test_final.hdf')
```

```python
test_df
```

```python
test_df.insert(
loc=4,
column="deg_preds",
value=None)
```

```python
test_df
```

```python
test_df.reset_index(inplace=True)
test_df.drop(columns="index", inplace=True)
```

```python
test_df.to_hdf('/data/caitlin/cross_validation/k2/k2_test_final.hdf',
               key='df')
```

```python
deg_path = Path('/data/caitlin/cross_validation/k2/k2_deg/DATA/')
```

```python
for ix, row in test_df.iterrows():
    deg_pred = pd.read_csv(deg_path.joinpath(row["trial_id"], row["trial_id"]).with_name(f'{row["trial_id"]}_predictions.csv'))
    deg_pred.drop(columns=["background", "Unnamed: 0"], inplace=True)
    pred_etho = deg_pred.to_numpy()
    test_df.loc[:, "deg_preds"].loc[ix] = pred_etho.T
```
