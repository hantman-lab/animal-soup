```python
import pandas as pd
import os
import shutil
from pathlib import Path
import random
```

```python
ground = pd.read_hdf('/data/caitlin/ground_truth/ground_truth_table_1000.hdf')
```

```python
ground
```

```python
# get a random sample of 700 indices for training split, remainder in testing
ixs = list(range(0, 1000))
training_ixs = random.sample(ixs, 700)
```

```python
testing_ixs = list(set(ixs) - set(training_ixs))
```

```python
k_train = ground[ground.index.isin(training_ixs)]
k_train
```

```python
k_test = ground[ground.index.isin(testing_ixs)]
k_test
```

```python
k_train.to_hdf('/data/caitlin/cross_validation/k1/k1_train.hdf', key='df')
k_test.to_hdf('/data/caitlin/cross_validation/k1/k1_test.hdf', key='df')
```

```python
k_train = pd.read_hdf('/data/caitlin/cross_validation/k1/k1_train.hdf')
k_test = pd.read_hdf('/data/caitlin/cross_validation/k1/k1_test.hdf')
```

```python
src = Path('/data/caitlin/ground_truth/')
dst = Path('/data/caitlin/cross_validation/k1/train_vids/')
```

```python
for ix, row in k_train.iterrows():
    shutil.copy(src.joinpath(row["animal_id"], row["session_id"], row["trial_id"]).with_suffix('.avi'),
                dst.joinpath(row["trial_id"]).with_suffix('.avi'))
```

```python
dst = Path('/data/caitlin/cross_validation/k1/test_vids/')
```

```python
for ix, row in k_test.iterrows():
    shutil.copy(src.joinpath(row["animal_id"], row["session_id"], row["trial_id"]).with_suffix('.avi'),
                dst.joinpath(row["trial_id"]).with_suffix('.avi'))
```
